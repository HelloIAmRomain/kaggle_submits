```python
# Import libraries
import pandas as pd
from copy import deepcopy
# Machine Learning Libraries with keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Importing the dataset
train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')
# Drop the columns that are not needed
train_data = train_data.drop(['Name','Ticket','Cabin'], axis=1)
test_data = test_data.drop(['Name','Ticket','Cabin'], axis=1)
```

```python
train = deepcopy(train_data)
test = deepcopy(test_data)
# Replace sex "male" and "female" with "0" and "1" respectively
train["Sex"] = train_data["Sex"].replace({"female":1, "male":0})
test["Sex"] = test_data["Sex"].replace({"female":1, "male":0})
train["Embarked"] = train_data["Embarked"].replace({"S":0, "Q":1, "C":2})
test["Embarked"] = train_data["Embarked"].replace({"S":0, "Q":1, "C":2})
print(train.head())
print("\n")
print(test.head())
```

```python
train_x, train_y = train.drop(['Survived'], axis=1), train['Survived']
print(train_x.head())
print("\n")
print(train_y.head())
```

```python
# Filling the missing values
train_x = train_x.fillna(train_x.mean())
test = test.fillna(test.mean())
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train_x_scaled = sc.fit_transform(train_x)
test_scaled = sc.transform(test)
print(train_x_scaled[:5])
```

```python
# Create model
input_params = train_x_scaled.shape[1]
model = Sequential()
model.add(Dense(32, input_dim=input_params, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
# Fit the model
model.fit(train_x_scaled, train_y, epochs=500, batch_size=10)
accuracy = model.evaluate(train_x_scaled, train_y)
print("\n%s: %.2f%%" % (model.metrics_names[1], accuracy[1]*100))
```

```python
# Use the model to predict the test dataset
predictions = model.predict(test_scaled)
predictions = [1 if x>=0.5 else 0 for x in predictions]
print(predictions[:10])
```

```python
# Save the predictions to a csv file
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submissions/my_submission.csv', index=False)
print("Your submission was successfully saved!")
```
